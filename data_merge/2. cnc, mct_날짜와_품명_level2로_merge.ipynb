1. 월과 일을 -로 바꾸자

```python
def filtering(x):
    x = x.replace("월", "-")
    x = x.replace("일", "")
    return x
```

2. 년 붙이기

```python
def plusYear(x):
    return str(20)+"-"+x
```

```python
def plusYear21(x):
    return str(21)+"-"+x
```

```python
def plusYear22(x):
    return str(22)+"-"+x
```

```python
df["날짜_"] = df["날짜_"].astype(dtype = "str")
```

```python
type(df["날짜_"][0])
```

```python
df["날짜_"] = df["날짜_"].apply(filtering)
```

```python
df20 = df.iloc[:1635]
df20
```
