```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

freq_factory = apriori(df, min_support=0.01, use_colnames=True)
# 0.01의 의미는 총 10885건의 제품 중에서 최소한 108건의 제품 해당 항목을 고려했을 때만 
# 해당 항목이 존재해야 한다는 것을 의미
```

```python
rules = association_rules(freq_factory, metric="lift", min_threshold=1)
rules.head()
```

```python
rules.info()
```

```python
rules["antecedents_len"] = rules["antecedents"].apply(lambda x: len(x))
rules
```

```python
# 지지도 5% 이상
rules[rules['antecedent support'] >= 0.05] 
```

```python
# 필요한 값들만 선별하여 결과표 다시 생성
rules = rules[['antecedents', 'consequents','antecedent support','consequent support',
                  'support', 'confidence', 'lift', 'antecedents_len']]
rules = rules.sort_values('lift', ascending = True)

rules.head()
```

```python
rules[rules['lift'] >= 1.0]
```

```python
# if 조건절의 지지도가 5% 이상이고, 향상도가 1보다 크며, 신뢰도가 어느 정도 큰 것을 보자
# 지지도가 5% 이상이면서 신뢰도가 큰 것부터 나열
pd.set_option('display.max_rows', None)

min_sup = rules[rules['antecedent support'] >= 0.05 & rules['support'] >= 0.2]
# min_sup = min_sup.sort_values('lift', ascending = False)
# min_sup
min_sup.sort_values('confidence', ascending = False)

# 450 -> 10% / 29% / 같이 9% / 99% / 3.3 / 3건
# 2 / 14 / 59 / 0 = 1 / 61 = 62 / 48 = 49 / 47 = 46 / 41 = 40 / 38 = 39 
# 14 = 15 / 
```

보통 최소 지지도 값 이상의 연관규칙들 중에서 큰 신뢰도 값을 갖는 규칙들이 의미 있는것으로 최종 판정된다

만약 향상도 값이 1 이 되면 X 와 Y 는 서로 독립이 된다. 즉, X 와 Y 는 서로 전혀 영향을 미치지 않는다는 것을 알 수 있다.

만약 향상도 값이 1 보다 크면 X 와 Y 는 서로 보완 관계에 있게 된다. 즉, X가 증가(감소)함에 따라 Y도 함께 증가(감소)한다. 반면에 X와 Y의 향상도가 1 보다 작게 되면 상호 대체 관계가 있게 된다. 즉 X 가 증가(감소)함에 따라 Y 는 감소(증가)하게 된다.


1에 가까울수록 같은 트랜잭션에 속한 X 와 Y 는 전체 트랜잭션에서 차지하는 중요도가 커지게 된다. 일반적으로 지지도는 자주 발생하지 않은 규칙을 우선적으로 제거하는데 사용한다. 지지도가 0 에 가깝다는 것은 전체 트랜잭션에서 차지하는 비중이 크지 않다는 것을 의미하기 때문이다.


다음 식과 같이 지지도는 (𝑋𝑋 → 𝑌𝑌) 와 (𝑌𝑌 → 𝑋𝑋) 의 차이를 알 수 없기 때문에 두 규칙 간의 차이를 알 수 없는 단점이 있다. 

antecedent support : 해당 공정이 일어난 횟수 / 10885 -> 선행(if)

consequent support : 해당 공정이 일어난 횟수 / 10885 -> 후행(then)

support(지지도) : 

전체 거래에서 특정 물품 A와 B가 동시에 거래되는 비중

해당 규칙이 얼마나 의미있는지 보여줌.

<br>
 

지지도 = P(A∩B) : 

A와 B가 동시에 일어난 횟수 / 전체 거래 횟수

<br>
 
confidence(신뢰도) : 

A를 포함하는 거래 중 A와 B가 동시에 거래되는 비중

<br>

신뢰도 = P(A∩B) / P(A) : 

A와 B가 동시에 일어난 횟수 / A가 일어난 횟수

 
<br>
 

lift(향상도) : 

A라는 상품에서 신뢰도가 동일한 상품 B와 C가 존재할 때, 어떤 상품을 더 추천해야 좋을지 판단.

A와 B가 동시에 거래된 비중을 A와 B가 서로 독립된 사건일 때 동시에 거래된 비중으로 나눈 값

 <br>

향상도 = P(A∩B) / P(A)*P(B) = P (B|A) / P (B) : 

A와 B가 동시에 일어난 횟수 / A, B가 독립된 사건일 때 A,B가 동시에 일어날 확률
