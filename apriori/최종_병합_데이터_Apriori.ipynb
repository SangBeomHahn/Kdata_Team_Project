```python
dfSub1 = df.drop("품명 코드", axis = 1)
dfSub1
```

```python
dfSub1 = dfSub1.set_index("품명")
dfSub1
```

```python
# 4403개의 제품 제조 건수에 대하여 평균적으로 cnc(50%), mct(49%), 세정(63%), 수입검사2(10%), 용점(10%), 품질검사(20%), 후처리(20%)
# IPA, LEAK 구동, 외주가공, 타각은 거의 안 한다.
dfSub1.describe()
```

```python
# 0.01의 의미는 총 4403건의 거래 중에서 최소한 40건의 거래 해당 항목을 고려했을 때만 해당 항목이 존재해야 한다는 것을 의미

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

freq_factory = apriori(dfSub1, min_support=0.01, use_colnames=True)
```

```python
rules = association_rules(freq_factory, metric="lift", min_threshold=1)
rules
```

```python
rules.info()
```

```python
rules["antecedents_len"] = rules["antecedents"].apply(lambda x: len(x))
rules
```

```python
# 필요한 값들만 선별하여 결과표 다시 생성
rules = rules[['antecedents', 'consequents','antecedent support','consequent support',
                  'support', 'confidence', 'lift', 'antecedents_len']]
rules = rules.sort_values('lift', ascending = True)
rules
```

# table 규칙
1. confidence 0.5 이상

2. support는 신뢰도보다 작거나 같다.(선후관계를 보려면 support를 봐야하는데 그게 신뢰를 가지려면 lift가 1보다 커야한다.), 어느 정도 커야한다. 어느 정도를 잡는 게 중요하다.

3. lift가 상대적으로 큰 건 배제

```python
df1 = rules[rules['lift'] >= 1.0]
df1
```

```python
df2 = df1[df1["confidence"] >= 0.5]
df2
```

```python
# 근데 애초에 결과가 그렇게 나옴
df2[df2["support"] > df2["confidence"]]
```

```python
# lift가 상대적으로 높은 건 너무 당연한 것이라서 빼는 것이 중요하다.
df2.sort_values('lift', ascending = False)
```

```python
# 그 상대적으로 높은 것을 9라고 하자 왜냐? 9이상의 값들은 수입검사, 구동검사, LEAK, 타각 및 포장 등 제품 후처리에 꼭 필요한 공정들이라 반드시 거치므로 lift 확률이 클 수 밖에 없다. = describe에서 비율이 낮게 나온 애들임
df3 = df2[df2["lift"] < 9].sort_values("lift", ascending = False)
df3
```

# 이제 support의 어느정도를 정해보자

```python
# support가 어느 정도 높아야한다고 했는데 그 어느 정도를 (최소 지지도 2)이라고 하자 -> 근데 0.01도 다 중요한 공정이라 빼기 어려움. 이대로 표를 만들자
df3.sort_values("support", ascending=False)
```
